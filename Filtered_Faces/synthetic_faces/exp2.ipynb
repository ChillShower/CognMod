```python
import pandas as pd
import matplotlib.pyplot as plt

ratings_amalie = pd.read_csv('image_reviews_amalie.csv')
ratings_simone = pd.read_csv('image_reviews_simone.csv')

ratings_amalie['true_rating'] = ratings_amalie['image'].str.extract(r"class_([0-9.]+)_weightmap.jpg").astype(float)
ratings_amalie['grade'] = ratings_amalie['grade'].astype(float)

ratings_simone['true_rating'] = ratings_simone['image'].str.extract(r"class_([0-9.]+)_weightmap.jpg").astype(float)
ratings_simone['grade'] = ratings_simone['grade'].astype(float)
```

```python
ratings_amalie
```

```python
res_a = ratings_amalie['grade'].corr(ratings_amalie['true_rating'], method='spearman')
print(res_a)
```

```python
ratings_simone
```

```python
res_s = ratings_simone['true_rating'].corr(ratings_simone['grade'], method='spearman')
print(res_s)
```

```python
groups1 = [ratings_amalie[ratings_amalie['grade'] == p]['true_rating'].values for p in sorted(ratings_amalie['grade'].unique())]

plt.figure(figsize=(7,5))
plt.boxplot(groups1, labels=sorted(ratings_amalie['grade'].unique()))
plt.xlabel("Predicted rating")
plt.ylabel("True rating")
plt.title("Participant 1")
plt.ylim(-1.0, 5.2)  # adjust if your rating scale is different
plt.show()
```

```python
groups = [ratings_simone[ratings_simone['grade'] == p]['true_rating'].values for p in sorted(ratings_simone['grade'].unique())]

plt.figure(figsize=(7,5))
plt.boxplot(groups, labels=sorted(ratings_simone['grade'].unique()))
plt.xlabel("Predicted rating")
plt.ylabel("True rating")
plt.title("Participant 2")
plt.ylim(-1.0, 5.2)  # adjust if your rating scale is different
plt.show()
```
