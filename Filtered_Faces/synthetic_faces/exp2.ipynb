```python
import pandas as pd
import matplotlib.pyplot as plt

ratings_amalie = pd.read_csv('image_reviews_amalie.csv')
ratings_simone = pd.read_csv('image_reviews_simone.csv')
ratings_alex = pd.read_csv('image_reviews_alex.csv')

# Participant 1
ratings_amalie['true_rating'] = ratings_amalie['image'].str.extract(r"class_([0-9.]+)_weightmap.jpg").astype(float)
ratings_amalie['grade'] = ratings_amalie['grade'].astype(float)

# Participant 2
ratings_simone['true_rating'] = ratings_simone['image'].str.extract(r"class_([0-9.]+)_weightmap.jpg").astype(float)
ratings_simone['grade'] = ratings_simone['grade'].astype(float)

# Participant 3
ratings_alex['true_rating'] = ratings_alex['image'].str.extract(r"class_([0-9.]+)_weightmap.jpg").astype(float)
ratings_alex['grade'] = ratings_alex['grade'].astype(float)
```

```python
res_amalie = ratings_amalie['grade'].corr(ratings_amalie['true_rating'], method='spearman')
print(res_amalie)
```

```python
res_simone = ratings_simone['grade'].corr(ratings_simone['true_rating'], method='spearman')
print(res_simone)
```

```python
res_alex = ratings_alex['grade'].corr(ratings_alex['true_rating'], method='spearman')
print(res_alex)
```

```python
groups_amalie = [ratings_amalie[ratings_amalie['grade'] == p]['true_rating'].values for p in sorted(ratings_amalie['grade'].unique())]

plt.figure(figsize=(7,5))
plt.boxplot(groups_amalie, labels=sorted(ratings_amalie['grade'].unique()))
plt.xlabel("Predicted rating")
plt.ylabel("True rating")
plt.title("Participant 1")
plt.ylim(0.0, 5.0)  # adjust if your rating scale is different
plt.show()
```

```python
groups_simone = [ratings_simone[ratings_simone['grade'] == p]['true_rating'].values for p in sorted(ratings_simone['grade'].unique())]

plt.figure(figsize=(7,5))
plt.boxplot(groups_simone, labels=sorted(ratings_simone['grade'].unique()))
plt.xlabel("Predicted rating")
plt.ylabel("True rating")
plt.title("Participant 2")
plt.ylim(0.0, 6.0)  # adjust if your rating scale is different
plt.show()
```

```python
groups_alex = [ratings_alex[ratings_alex['grade'] == p]['true_rating'].values for p in sorted(ratings_alex['grade'].unique())]

plt.figure(figsize=(7,5))
plt.boxplot(groups_alex, labels=sorted(ratings_alex['grade'].unique()))
plt.xlabel("Predicted rating")
plt.ylabel("True rating")
plt.title("Participant 3")
plt.ylim(0.0, 5.0)  # adjust if your rating scale is different
plt.show()
```
